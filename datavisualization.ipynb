```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
sns.get_dataset_names()
```

```python
df=sns.load_dataset('tips')
```

```python
df
```

```python
custom_bins=[0,10,20,30,40,50,60]
sns.histplot(data=df,x="total_bill",bins=custom_bins)
plt.show()
```

```python
sns.histplot(data=df,x='total_bill',kde=True,stat='probability')
```

```python
sns.scatterplot(data=df,x="total_bill",y="tip")
plt.show()
```

```python
sns.barplot(x="day",y="tip",data=df)
plt.xlabel("day")
plt.ylabel("Tip")
plt.title("Relationship between total bil and tips") 
plt.show()
```

```python
sns.barplot(x="day",y="tip",data=df,ci=None)
plt.xlabel("day")
plt.ylabel("Tip")
plt.title("Relationship between total bil and tips") 
plt.show()
```

```python
sns.barplot(x="day",y="tip",data=df,ci=None,palette='pastel')
plt.xlabel("day")
plt.ylabel("Tip")
plt.title("Relationship between total bil and tips") 
plt.show()
```

```python
sns.barplot(x="day",y="tip",data=df,ci=None,palette='pastel',saturation=0.1)
plt.xlabel("day")
plt.ylabel("Tip")
plt.title("Relationship between total bil and tips") 
plt.show()
```

 #### a parameter in seaborn plot functions that allow you to specify a categorical variable.

#### box plot(sns.boxplot):summarize the distributionof the numeric variable

```python
sns.boxplot(x='day',y='total_bill',data=df)
plt.show()
```

#### outliers are data points that are significally different from the rest of the data,they lie away from the group of main points

#MINIMUM:THE SMALLEST VALUE IN THE DATASET
#FIRST QUARTILE:THE MEADIAN OF THE LOWER HALFOF THE DATA(EXCLUDING THE MEADIAN IF THE MEMBER OF OBSERVATION IS ODD)


```python
sns.boxplot(x='day',y='total_bill',data=df,hue='size')
plt.show()
```

```python
sns.boxplot(x="day",y="total_bill",data=df,hue="size",linewidth=1)
plt.show()
```

```python
sns.boxplot(x='day',y='total_bill',data=df,linewidth=1,showmeans=True)
plt.show()
```

```python
sns.boxplot(x='day',y='total_bill',data=df,linewidth=1,showmeans=True,meanprops={"marker": "+","mec":"black","markersize":10})
plt.show()
```

# correlation

```python
#it me9ns the relationshio between two numerical variable
df.corr()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
for column in ['day','time','sex','smoker']:
    df[column]=le.fit_transform(df[column])
df
```

```python
corr=df.corr()
corr
```

```python
sns.heatmap(corr)
```

```python
sns.heatmap(corr,annot=True)
```

```python
corr.isna().sum()
```
